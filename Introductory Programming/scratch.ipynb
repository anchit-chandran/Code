```python
import math

def solution_2()->list[int]:
    
    # Constants
    X = 70
    H = 25
    
    nums = input('Please enter 3 integers, separated by commas, without whitespace: ')  
    
    Y = [int(num) for num in nums.split(',')]
    
    output = []
    
    for y in Y:
        numerator = 3 * X * y 
        denominator = H
        
        answer = round(math.sqrt((numerator / denominator)))
        
        output.append(answer)
    
    return output

print(solution_2())
```

```python
import re
from operator import itemgetter

def get_input()->tuple[str, int, int]:
    
    player_input = input('Please enter all players\' data in the following format:\n\n(\'NAME\', AGE, SCORE)\n\nReplace capitalised words with the actual value and separate each item with a comma\n\ne.g. (\'Anchit\', 25, 10),(\'Anchit\', 25, 10)')

    data = []
    regex_pattern = r"\('\w+',\d+,\d+\d\)"
    for item in re.findall(regex_pattern, player_input):
        item_cleaned = item.replace('(','').replace(')','').replace('\'','').split(',')
        data.append(
            (item_cleaned[0], int(item_cleaned[1]), int(item_cleaned[2]))
        )
    
    return data
    

def solution_3(player_data):
    sorted_data = sorted(player_data, key=itemgetter(0,1,2))
    return sorted_data

# ('John',20,80),('Jony',19,95),('Anna',18,70),('Json',23,120)
player_data = get_input()
print(solution_3(player_data=player_data))
```

```python
def solution_4(list_a:list, list_b:list):
    
    list_a, list_b = set(list_a), set(list_b)
    
    unique_common_elements = list_a.intersection(list_b)
    
    return list(unique_common_elements)

list_a = [5, 23, 7, 89,7, 90]
list_b = [89, 25, 89, 7, 14, 76, 5, 11, 5]

print(solution_4(list_a, list_b))
```

```python
def solution_5(candidate_string):
    start, end = 0, -1

    for _ in range(len(candidate_string) // 2):
        if candidate_string[start] != candidate_string[end]:
            print("Not a palindrome")
            return False

        start += 1
        end -= 1

    print("This is a palindrome")
    return True


print(solution_5("racecar"))
assert solution_5("racecar") == True
```

```python
class Board:
    def __init__(self):
        self.squares = {num: None for num in range(1, 10)}
        self.player_turn = True # player starts first

    def _get_rows(self) -> list[list]:
        sqrs = list(self.squares.values())

        list_of_squares = []

        # grab squares in batches of 3
        for i in range(0, len(sqrs), 3):
            list_of_squares.append(sqrs[i : i + 3])

        return list_of_squares

    def _set_square_value(self, square: int, value:str):
        print(f"updating square {square}")
        self.squares[square] = value

    def make_move(self, square: str):
        square = int(square)
        
        value = 'O' if self.player_turn else 'X'
        
        # update turn state
        self.player_turn = not self.player_turn
        
        self._set_square_value(square=square, value=value)

    def _render_row(self, row_num:int):
        
        row = self._get_rows()[row_num]
        print(row)
    
    def display_board(self) -> None:
        """Renders and prints the current board"""

        for j in range(len(self._get_rows())):
            stringified_row = ""
            row = self._get_rows()[j]

            # render each row
            for i in range(len(row)):
                val = row[i]
                val = val if val is not None else ""
                sep = "|" if i < 2 else ""
                stringified_row += f" {val} {sep}"

            # add row separation line
            if j < 2:
                stringified_row += "\n" + '-'*9

            print(stringified_row)


def solution_7():
    board = Board()

    board.display_board()

    for i in range(9):
        board.make_move(i)
        board.display_board()


solution_7()
```
